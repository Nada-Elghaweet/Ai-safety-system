```python
import os
os.listdir("/content")
```

```python
path = "/content/Processed_NON_VIOLENCE_dataset"
print ( "images found" , len(os.listdir(path) ))
```

```python
import shutil
from sklearn.model_selection import train_test_split
```

```python
output_path = "/content/split_dataset"
train_dir = os.path.join(output_path, "train")
val_dir = os.path.join(output_path, "val")
test_dir = os.path.join(output_path, "test")
```

```python
for folder in [train_dir, val_dir, test_dir]:
    os.makedirs(folder, exist_ok=True)
```

```python
all_images = [os.path.join(path, f) for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
```

```python
train_imgs, temp_imgs = train_test_split(all_images, test_size=0.2, random_state=42)
val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.2, random_state=42)
```

```python
def copy_files(file_list, target_folder):
    for f in file_list:
        shutil.copy(f, target_folder)

# Copy the splits
copy_files(train_imgs, train_dir)
copy_files(val_imgs, val_dir)
copy_files(test_imgs, test_dir)

print(f"Train: {len(train_imgs)} | Val: {len(val_imgs)} | Test: {len(test_imgs)}")
```

```python
zip_output = "/content/split_dataset.zip"
shutil.make_archive("/content/split_dataset", 'zip', output_path)
```
