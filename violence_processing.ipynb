```python
import os
os.listdir("/content")
```

```python
path = "/content/Processed_fight_images_dataset"
print("images found", len(os.listdir(path)))
```

```python
from PIL import Image , UnidentifiedImageError
import numpy as np
import shutil
from sklearn.model_selection import train_test_split
import hashlib
```

```python
processed_path = "/content/fight_images_processed"
img_size = (224 , 224) #suitable for cnn
os.makedirs(processed_path, exist_ok=True)
```

```python
#corrupted images skipping

valid_imgs= []
for file_name in os.listdir(path) :
  src_path = os.path.join(path , file_name)
  try:
    with Image.open(src_path) as img:
      img = img.resize(img_size)
      img = img.convert("RGB")
      dst_file_name = os.path.splitext(file_name)[0] + ".jpg"
      dst_path = os.path.join(processed_path , dst_file_name)
      img.save(dst_path , format = "JPEG")
      valid_imgs.append(dst_path)
  except UnidentifiedImageError:
    print (f"corrupted image skipped : {file_name}")
print(f"total valid images : {len(valid_imgs)}")
```

```python
from enum import unique
#removing duplicates

hashes = set()
unique_imgs = []
for img_path in valid_imgs:
  with open(img_path , "rb") as f:
    h = hashlib.md5(f.read()).hexdigest()
  if h not in hashes:
      hashes.add(h)
      unique_imgs.append(img_path)
  else:
    os.remove(img_path)
    print(f"duplicate image removed : {img_path}")

print(f"total unique images : {len(unique_imgs)}")
```

```python
train_val, test = train_test_split(unique_imgs, test_size=0.2, random_state=42)
train, val = train_test_split(train_val, test_size=0.2, random_state=42)
```

```python
print(f"Train images: {len(train)}")
print(f"Validation images: {len(val)}")
print(f"Test images: {len(test)}")
```

```python
def copy_files (file_list  , folder_name):
  folder_path = os.path.join(processed_path , folder_name)
  os.makedirs(folder_path , exist_ok= True )
  for f in file_list:
    shutil.copy(f, folder_path)

copy_files(train , "train")
copy_files(val , "val")
copy_files(test , "test")
```
